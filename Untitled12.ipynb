```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = r"C:\Users\aksha\OneDrive\Documents\gym_membership.xlsx"
df = pd.read_excel(file_path)

# Clean column names
df.columns = df.columns.str.strip()

# Print available columns to check for 'Age'
print("Available columns:", df.columns)

# Check if 'Age' column exists
if 'age' in df.columns:
    # Create a histogram to show the distribution of Age
    plt.figure(figsize=(8, 6))
    plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black')
    plt.title('age Distribution of Population')
    plt.xlabel('age')
    plt.ylabel('Number of People')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Categorize people into different age groups
    age_bins = [0, 18, 30, 45, 60, 100]
    age_labels = ['0-18', '19-30', '31-45', '46-60', '61+']
    df['Age Group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

    # Count the number of people in each age group
    age_group_counts = df['Age Group'].value_counts()

    # Print the number of people in each age group
    print("Number of people in each age group:")
    print(age_group_counts)

    # Plot the age group counts
    plt.figure(figsize=(6, 4))
    age_group_counts.plot(kind='bar', color='lightcoral')
    plt.title('Number of People in Each Age Group')
    plt.xlabel('Age Group')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
else:
    print("The 'age' column is missing in the DataFrame.")
```
